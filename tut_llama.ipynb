```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ro")
```

```python
def build_prompt(src="", pairs=[("", "")]):
  pairs_text = [f"-{src}: {dest}" for (src,dest) in enumerate(pairs)]

  return f"Translate the text delimited by ------- from English to Romanian. Reply only the translated text." + \
         "Use these examples of already translated words and expressions:" + \
         "\n-------\n" + \
         "\n" + "\n".join(pairs_text) + \
         "\n-------\n" + \
         "\n" +f"{src}"
```

```python
test_text = """
Ana has apples
"""
test_pairs = [
    ("corn", "cucuruz"),
    ("hammer mill", "moară cu ciocane"),
    ]

test_prompt = build_prompt(test_text, [("")])
```
